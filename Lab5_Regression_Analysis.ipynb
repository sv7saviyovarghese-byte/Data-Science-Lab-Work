```python
X = mydata.drop(columns = ["MEDV"])
y = mydata["MEDV"]

```

```python
X.shape
X.columns
```

```python
y.shape
```

```python
###Step 5: divide the data intro train and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
##Step 6: Applying machine leraning algorithms

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr_model = lr.fit(X_train, y_train)

```

```python
lr_model.coef_
```

```python
lr_model.intercept_
```

```python
y_pred_lr = lr.predict(X_test)
```

```python
from sklearn.metrics import root_mean_squared_error as rmse
rmse_lr = rmse(y_test, y_pred_lr)
rmse_lr 
```

```python
from sklearn.metrics import r2_score as r2

r2_lr = r2(y_test, y_pred_lr)
r2_lr
```

```python
#### Decision tree regressor 

from sklearn.tree import DecisionTreeRegressor

dtr = DecisionTreeRegressor()
lr_model = dtr.fit(X_train, y_train)


y_pred_dtr = dtr.predict(X_test)
rmse_dtr = rmse(y_test, y_pred_dtr)
r2_dtr = r2(y_test, y_pred_dtr)
```
